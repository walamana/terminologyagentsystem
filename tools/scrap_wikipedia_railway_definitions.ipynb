```python
import requests as r
from bs4 import BeautifulSoup

res = r.get("https://de.wikipedia.org/w/index.php?title=Liste_von_Abk%C3%BCrzungen_im_Eisenbahnwesen&oldid=256152030").text

soup = BeautifulSoup(res, 'lxml')
```

```python
tables = soup.find_all('table')

entries = []
for table in tables:
    for row in table.find_all_next("tr"):
        parts = row.find_all_next("td")
        entries.append((parts[0].text.strip(), parts[1].text.strip()))
```

```python
import pandas as pd
pd.DataFrame(entries).to_csv("../data/common_abbreviations.csv", index=True)
```

```python
from collections import defaultdict

grouped = defaultdict(list)
for key, value in entries:
    grouped[key].append(value)

len([abbreviation for abbreviation, terms in grouped.items() if len(terms) > 1])
```
